```python
# Setup
t = np.linspace(0,10,1000)
f1 = 0.5 # Low frequency signal
x1 = np.cos(2*np.pi*f1*t)
plt.figure
plt.plot(t, x1, 'b')
```

```python
f2 =  5
xr = ss.rect(t-4.5,0.5)
x2 = x1 + 0.3*xr*np.cos(2*np.pi*f2*t)
plt.figure
plt.plot(t, x2, 'b')
```

# Filter design

```python
fs = 100
f_stop = 1
f_pass = 2
b_but,a_but,sos_but = iir_d.IIR_hpf(f_stop,f_pass,0.5,60,fs,'butter')
```

#Simulation

```python
x2f = signal.filtfilt(b_but, a_but, x2) # Apply filter to x2
tv = np.where(np.abs(x2f) > 0.2)
fig, ax = plt.subplots()
ax.add_patch(Rectangle((tv[0][0]-1, -1), 100, 2,color="mistyrose"))
ax.plot(x2)

```
