```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from scipy import signal
import random
```

```python
# detect-sub-series
def detect_sub(series, subseries,threshold):
    #return time first occurrence
    series = np.array(series)
    subseries = np.array(subseries)
    L = series.shape[0]
    M = subseries.shape[0]
    mins = np.min(series)
    maxs = np.max(series)
    seriesn = (series - mins)/(maxs - mins)
    subseriesn = (subseries - mins)/(maxs - mins)
    nocur = 0
    k = 0
    flag = False
    while (flag == False) & (k < L-M):
        ew = seriesn[k:k+M] - subseriesn
        if np.linalg.norm(ew) < threshold:
            flag = True
        k = k +1
    return k
```

```python
random.seed(0)
x = [2,6,6,5,3] # pattern
X = []
for i in range(0, 50):
    # any random numbers from 0 to 1000
    X.append(random.randint(0, 10))

X = X + x #include pattern so that at least we have one occurrence

for i in range(0, 50):
    # any random numbers from 0 to 1000
    X.append(random.randint(0, 10))
```

```python
fig, ax = plt.subplots()
ax.plot(x)
ax.set_title('Pattern')
```

```python
td = detect_sub(X, x,0.1)
print(td)
```

```python
fig, ax = plt.subplots()
ax.add_patch(Rectangle((td-1, 0), len(x)-1, 10,color="mistyrose"))
ax.plot(X)
ax.set_title('Time series')
```
